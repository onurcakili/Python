```python
def karesini_al(x):
    return pow(x,2)

karesini_al(2)
```

```python
sayilar = [*range(1,6)]
sayilar
```

```python
for i in range(len(sayilar)):
    print(karesini_al(i))
```

```python
list(map(karesini_al, sayilar))
```

```python
def cift_sayilari_filtrele(x):
    if x % 2 == 0:
        return x
    else:
        return None

cift_sayilari_filtrele(4)
```

```python
def cift_sayilari_filtrele(x):
    return x if x % 2 == 0 else None

cift_sayilari_filtrele(4)
```

```python
sayilar = [*range(1,10)]
sayilar
```

```python
list(filter(cift_sayilari_filtrele, sayilar))
```

```python
# kare alan lambda

list(map(lambda sayi : sayi ** 2, sayilar))
```

```python
# cift sayilari listeleyen lambda

list(filter(lambda sayi : sayi % 2 == 0, sayilar))
```

```python
list(filter(lambda x : x if x % 2 == 0 else None, sayilar ))
```
