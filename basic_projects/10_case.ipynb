```python
class Hayvan:
    def __init__(self, isim):
        self.isim = isim
        
    def ses_cikar(self):
        return "..."
        
class Kedi(Hayvan):
    def ses_cikar(self):
        return "Miyav!"
    
class Kopek(Hayvan):
    def ses_cikar(self):
        return "Hav Hav!"

kedi = Kedi("Tekir")
kopek = Kopek("Sevimli")

print(f"{kedi.isim}: {kedi.ses_cikar()}")    # Tekir: Miyav!
print(f"{kopek.isim}: {kopek.ses_cikar()}")  # Karabaş: Hav hav!
```
