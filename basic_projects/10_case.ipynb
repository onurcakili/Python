```python
def tam_isim(isim, soyisim):
    return f"{isim} {soyisim}".upper()

tam_isim("onur","cakili")


# upper - > tamamını büyük yapar
# title - > baş harfleri büyük yapar
# capitalize -> sadece ilk kelimenin baş harfini büyük yapar
```

```python
def filtrele(sayilar, esik_deger):
    sonuc = []   # yer tutucu
     
    for sayi in sayilar:
        if sayi > esik_deger:
            sonuc.append(sayi)    # listeye eklemek 
    return sonuc # döng. dısında return tanımla

filtrele([1,2,3,4,5],3)   # def, liste olarak arg. alabilir
```

```python
def ogrenci_kart(**ogrenci):
    
    for key, value in ogrenci.items():
        print(f"{key}: {value}")
    if "yas" in ogrenci and ogrenci["yas"] < 18:
        print("⚠️ Reşit değil")

ogrenci_kart(isim="Ayşe", yas=16, sehir="Ankara")
```

```python
#   *args = degerleri tuple olarak tutar
#   *kwargs = degerleri dict olarak tutar

def cift_topla(*sayilar):
    sonuc = 0 
    
    for sayi in sayilar:
        if sayi % 2 == 0:
            sonuc += sayi
    return sonuc
    
    
cift_topla(1,2,3,4,5,6)
```

```python
class Hayvan:
    def __init__(self, isim):
        self.isim = isim
        
    def ses_cikar(self):
        return "..."
        
class Kedi(Hayvan):
    def ses_cikar(self):
        return "Miyav!"
    
class Kopek(Hayvan):
    def ses_cikar(self):
        return "Hav Hav!"

kedi = Kedi("Tekir")
kopek = Kopek("Sevimli")

print(f"{kedi.isim}: {kedi.ses_cikar()}")    # Tekir: Miyav!
print(f"{kopek.isim}: {kopek.ses_cikar()}")  # Karabaş: Hav hav!
```
