```python
def filtrele(sayilar, esik_deger):
    sonuc = []   # yer tutucu
     
    for sayi in sayilar:
        if sayi > esik_deger:
            sonuc.append(sayi)    # listeye eklemek 
    return sonuc # döng. dısında return tanımla

filtrele([1,2,3,4,5],3)   # def, liste olarak arg. alabilir
```

```python
#   *args = degerleri tuple olarak tutar
#   *kwargs = degerleri dict olarak tutar

def cift_topla(*sayilar):
    sonuc = 0 
    
    for sayi in sayilar:
        if sayi % 2 == 0:
            sonuc += sayi
    return sonuc
    
    
cift_topla(1,2,3,4,5,6)
```

```python
class Kisi():
    def __init__(self, isim, yas):
        self.isim = isim
        self.yas = yas

    def tanit(self):
        print(f"Merhaba, ben {self.isim}, {self.yas} yaşındayım.")

k = Kisi("Onur", 25)
k.tanit()
```

```python
class Dikdortgen():
    def __init__(self, en, boy):
        self.en = en
        self.boy = boy

    def alan(self):
        alan = self.en * self.boy
        return alan

    def cevre(self):
        cevre = 2 * (self.en + self.boy)
        return cevre

d = Dikdortgen(4, 5)
print(d.alan())    # 20
print(d.cevre())   # 18
```

```python
class Sepet():
    def __init__(self):
        self.urunler = []       # dışarıdan gelmiyor, boş liste ile başlıyoruz

    def ekle(self, urun):
        self.urunler.append(urun)

    def cikar(self, urun):
        if urun in self.urunler:
            self.urunler.remove(urun)
        else:
            print(f"{urun} sepette yok!")

    def listele(self):
        print(self.urunler)

sepet1 = Sepet()
sepet1.ekle("elma")
sepet1.ekle("süt")
sepet1.cikar("elma")
sepet1.listele()   # ["süt"]
```

```python
class Ogrenci:
    def __init__(self, isim, notlar):
        self.isim = isim
        self.notlar = notlar

    def ortalama(self):
        return sum(self.notlar) / len(self.notlar)

    def basari_durumu(self):
        if self.ortalama() >= 50:    # başka metodu içerde çağırabildik!
            return "Geçti"
        else:
            return "Kaldı"

ogrenci1 = Ogrenci("Ayşe", [70, 80, 90])
print(ogrenci1.ortalama())         # 80.0
print(ogrenci1.basari_durumu())    # Geçti
```

```python
class ATM:
    def __init__(self):
        self.bakiye = 0         # dışarıdan gelmiyor, 0 ile başlıyoruz

    def para_yatir(self, miktar):
        if miktar <= 0:
            print("Geçersiz işlem!")
        else:
            self.bakiye += miktar
            print(f"{miktar} TL yatırıldı.")

    def para_cek(self, miktar):
        if miktar <= 0:
            print("Geçersiz işlem!")
        elif miktar > self.bakiye:
            print("Yetersiz bakiye!")
        else:
            self.bakiye -= miktar
            print(f"{miktar} TL çekildi.")

    def bakiye_goster(self):
        print(f"Güncel bakiye: {self.bakiye} TL")

atm = ATM()
atm.para_yatir(1000)
atm.para_cek(200)
atm.bakiye_goster()    # Güncel bakiye: 800 TL
atm.para_cek(5000)     # Yetersiz bakiye!
atm.para_cek(-50)      # Geçersiz işlem!
```
