```python
def asal_mi(sayi):
    
    if sayi < 2:
        return False
    else:
        for i in range(2, sayi):    ## 2, int(math.sqrt(sayi)) + 1): 
            if sayi % i == 0:
                return False

    return True

print(asal_mi(7))    # True
print(asal_mi(10))   # False
print(asal_mi(2))    # True
print(asal_mi(1))    # False
    
```

```python
def ortak_elemanlar(elemanlar1, elemanlar2):
    ortak = []
    
    for eleman1 in elemanlar1:
        for eleman2 in elemanlar2:
            if eleman1 == eleman2 and eleman1 not in ortak:
                ortak.append(eleman1)
    
    return ortak   # print değil return!

print(ortak_elemanlar([1,2,2,3], [2,2,4]))   # [2]
print(ortak_elemanlar([1,2,3,4], [3,4,5,6])) # [3, 4]
```

```python
def kelime_sayac(kelime):
    
    kelimeler = {}
    
    for x in kelime.split():
        if x in kelimeler:
            kelimeler[x] += 1
        else:
            kelimeler[x] =  1
    
    return kelimeler

kelime_sayac("merhaba dunya merhaba")
```

```python
def istatistik(liste):
    sirali = sorted(liste)
    orta = len(sirali) // 2
    
    if len(sirali) % 2 == 0:
        medyan = (sirali[orta - 1] + sirali[orta]) / 2
    else:
        medyan = float(sirali[orta])
    
    sonuc = {
        "adet"     : len(liste),
        "toplam"   : sum(liste),
        "ortalama" : sum(liste) / len(liste),
        "en_buyuk" : max(liste),
        "en_kucuk" : min(liste),
        "medyan"   : medyan
    }
    return sonuc

print(istatistik([10, 20, 30, 40, 50]))
```

```python
class Dikdortgen:
    def __init__(self, en, boy):
        self.en = en
        self.boy = boy

    def alan(self):
        return self.en * self.boy

    def cevre(self):
        return 2 * (self.en + self.boy)

    def __str__(self):
        return f"{self.en}x{self.boy} dikdörtgen"

d = Dikdortgen(4, 5)
print(d.alan())    # 20
print(d.cevre())   # 18
print(d)           # 4x5 dikdörtgen
```

```python
class Oyuncu:
    def __init__(self, isim):
        self.isim = isim
        self.can = 100
        self.skor = 0

    def hasar_al(self, miktar):
        if self.can - miktar < 0:
            self.can = 0
            print("Can 0'ın altına düşemez, can: 0")
        else:
            self.can -= miktar

    def puan_kazan(self, miktar):
        self.skor += miktar

    def durum(self):
        print(f"{self.isim} | Can: {self.can} | Skor: {self.skor}")

o = Oyuncu("Onur")
o.hasar_al(30)
o.hasar_al(80)
o.puan_kazan(100)
o.durum()   # Onur | Can: 0 | Skor: 100
```

```python
class Kutuphane:
    def __init__(self):
        self.kitaplar = []

    def kitap_ekle(self, isim, kitap):
        yeni_kitap = {"isim": isim, "kitap": kitap}
        self.kitaplar.append(yeni_kitap)

    def kitap_sil(self, kitap):
        for k in self.kitaplar:
            if k["kitap"] == kitap:
                self.kitaplar.remove(k)
                return
        print("Böyle bir kitap yok!")

    def listele(self):
        for k in self.kitaplar:
            print(f"{k['kitap']} - {k['isim']}")

    def yazar_ara(self, yazar):
        sonuc = []
        for k in self.kitaplar:
            if k["isim"] == yazar:
                sonuc.append(k["kitap"])
        return sonuc

k = Kutuphane()
k.kitap_ekle("Dostoyevski", "Suç ve Ceza")
k.kitap_ekle("Dostoyevski", "Karamozov Kardeşler")
k.kitap_ekle("Sabahattin Ali", "Kürk Mantolu Madonna")
k.listele()
k.kitap_sil("Karamozov Kardeşler")
k.listele()
print(k.yazar_ara("Dostoyevski"))
```
